# Question 2: Alphabetical Sort

For order a text the algorithm uses the counting sort on the words encoded with a unique number

```python
from sorting_lib import counting_sort

counting_sort("we are alive")
```

The encoding part is the most difficult, since the counting sort works on numbers rather than strings.
So, for encode the words with a number that represent each word in the text univocally and gives us a clue about the relationship of bigger and smaller between the other words, first we had to find the maximum length of the word (so we can give at each first letter of the words the same weight).

The fist task was to get only lowercase (since uppercase has different ord() value to lowercase)
Then we had to evaluate each letter number value with the ord() function, that was a critical step since we had to minimize the number we got from the ord function, so was necessary to subtract the smaller letter ord()-1 value to each ord value (so we can start to count form 1 instead of ord(a)). Then we had to assign to each word a weight based on the position, using a too short number for this operation will lead us to have two or more strings represented by the same value, use a too big number instead guarantee the univocal representation of the numbers, but will cause to have a too big number as coefficient, with bad consequences on the dimensions of the counting sort vector, that will course more time required for the evaluation and more memory required as well.

So the optimal number come up as the ord() of the bigger character minus the value of the littler. That number raised by the length-1 of the longest string in the text was used as coefficient of the fist character's value of all the words of the text, number raised by the length-2 as the coefficient of the second character's value and so on..
at the end those values where passed to the counting sort algorithm.
However, the performance of the counting sort are not that good in this case, since the numbers passed are really long, them grows really fast with the length of the longest word for the text. Indeed, the counting sort is really good with a little range of values.

#### Running time
So the algorithm runs in O(n + m), where n is the number of words in the text and m is the maximum value assigned to the text's words. When the text is composed by short words it works well, even if the length of the text is massive, otherwise is pretty slow.

# K-Means Initialization Problem

The K-Means algorithm converges to local optimum of the cost function. The initialization is very 
important for this type of clustering. Cost of the solution can be larger than the cost of optimal solution.

Below is an example: Supposing that the data consists of n points in five clusters (of some tiny radius
δ) arranged in a line, with some large distance B between them:

![image](https://user-images.githubusercontent.com/43060519/69825644-471bc380-1210-11ea-8b0f-75617243e737.png)

The optimal 5-clustering has cost roughly δ
2n. If we initialize k-means by choosing five centers at random
from the data, there is some chance that we’d end up with no centers from cluster 1, two centers from cluster
3, and one center each from clusters 2, 4, and 5:

![image](https://user-images.githubusercontent.com/43060519/69825764-a548a680-1210-11ea-844e-1be49769f497.png)

In the first round of k-means, all points in clusters 1 and 2 will be assigned to the leftmost center. The
two centers in cluster 3 will end up sharing that cluster. And the centers in clusters 4 and 5 will move
roughly to the centers of those clusters.

![image](https://user-images.githubusercontent.com/43060519/69825787-babdd080-1210-11ea-9ca2-2d82b28f229f.png)

Thereafter, no further changes will occur. This local optimum has cost Ω(B2n). We can make this
arbitrarily far away from the optimum cost by setting B large enough. Thus, good initialization is crucial.
